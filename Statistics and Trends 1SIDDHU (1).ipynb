 # line graph

```python
import matplotlib.pyplot as plt
x = df["Year"]
y = df["Primary energy consumption per capita (kWh/person)"]
plt.scatter(x,y,s = 3)
plt.xlabel('years')  
plt.ylabel('PEC_per_capita')  # Add y-axis label
plt.title("PEC_per_capita over years")  # Add a title
plt.legend("PEC_per_capita",loc = "upper left")
plt.xticks(rotation=45)
plt.show()
```

```python

y = df["Primary energy consumption per capita (kWh/person)"]
x = df["Year"]
# Plot the histogram
plt.hist(x, bins=20, edgecolor='black')  # Adjust 'bins' based on data range

# Labeling the axes and title
plt.xlabel('year')  
plt.ylabel('PEC_per_capita')  # y-axis label is now Frequency
plt.title("Histogram of PEC_per_Capita")  # Add a title
plt.xticks(rotation=45)
plt.legend("PEC_per_capita")
plt.show()
```

# Violin plot

```python

x = df["Primary energy consumption per capita (kWh/person)"]


plt.violinplot(x, vert=False) 

plt.xlabel('GDP per Capita')  
plt.title("Violin plot of PEC_per_capita")  
plt.show()
```
